### 3. Thay thế bằng giá trị trung bình (median) của cột

```python
df = pd.read_csv('houseprice.csv')
x = df.median()
print(x)
df.fillna(x,inplace=True)
df
df.isna().sum()
```

### 4. Thay thế bằng giá trị xuất hiện thường xuyên nhất trong cột

```python
df = pd.read_csv('houseprice.csv')
x = df.mode().iloc[0]
print(type(x))
print(x)
df2 = df.fillna(x,inplace=False)

df2.isna().sum()
```
