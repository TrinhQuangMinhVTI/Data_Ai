```python
import numpy as np
x = np.random.randint(1,100,100)
print(x)
print(type(x))
```

```python
import matplotlib.pyplot as plt
plt.plot(x)
plt.title(' th gc x')
plt.show()
```

```python
def find_max(ds):
    max_value = ds[0]
    for num in ds:
        if num > max_value:
            max_value = num
    return max_value
list_data = x
max_value_data = find_max(list_data)
print('Max value in list: ', max_value_data)
```

```python
a = max(list_data)
print(a)
```

```python
min_value_data = min(list_data)
print(min_value_data)
```

```python
def normalization(list_data):
    normal_list = []
    for i in list_data :
        a = (i - min_value_data)/(max_value_data - min_value_data)
        normal_list.append(a)
    return normal_list
a = normalization(list_data)
print(a)
```

```python
list_data_nor = (list_data - min_value_data)/(max_value_data-min_value_data)
print(list_data_nor)
```

```python
#  th:
plt.plot(list_data_nor)
plt.title('X sau khi Normalization')
plt.show()
```

```python
#Tinh mean:

def mean_c1(listx):
    sum = 0
    for i in listx:
        sum += i
    meanx = sum/len(listx)
    return meanx

mean_c1(x)
```

```python
def mean_c2(listx):
    sum = np.sum(listx)
    meanx = sum/len(listx)
    return meanx

mean_c2(x)
```

```python
mean_c3 = np.mean(x)
print(mean_c3)
```

```python
# do lech chuan std standard:
def standard(data):
    size = np.size(data)
    y = np.sum((data - mean_c2(data))**2)
    z = y/size
    std = np.sqrt(z)

    return std

resuld = standard(x)
print(resuld)
```

```python
# tinh standardization

def Standardization(data):
    std_data = (data - mean_c2(data))/standard(data)
    return std_data

result1 = Standardization(x)
print(result1)
```

```python
plt.plot(result1)
plt.title('X sau khi standardization')
plt.show()
```
